```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
%matplotlib inline


data = pd.read_csv("files/data.csv")
data
```

```python
plt.figure(figsize=(14, 6))
fig, ax = plt.subplots()
# Plot the data
ax.plot(data["epoch"], data["loss"], label="Train Loss")
ax.plot(data["epoch"], data["val_loss"], label="Valid Loss")

plt.legend()
# Set the y-axis tick formatter to display only integer values
ax.xaxis.set_major_formatter(ticker.FormatStrFormatter('%d'))
plt.show();
```

```python
plt.figure(figsize=(14, 6))
fig, ax = plt.subplots()
# Plot the data
ax.plot(data["epoch"], data["iou"], label="Train IOU")
ax.plot(data["epoch"], data["val_iou"], label="Valid IOU")

plt.legend()
# Set the y-axis tick formatter to display only integer values
ax.xaxis.set_major_formatter(ticker.FormatStrFormatter('%d'))
plt.show();
```

```python
plt.figure(figsize=(14, 6))
fig, ax = plt.subplots()
# Plot the data
ax.plot(data["epoch"], data["precision"], label="Train Precision")
ax.plot(data["epoch"], data["val_precision"], label="Valid Precision")

plt.legend()
# Set the y-axis tick formatter to display only integer values
ax.xaxis.set_major_formatter(ticker.FormatStrFormatter('%d'))
plt.show();
```
